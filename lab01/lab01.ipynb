# 4. Primer ejercicio con datos

````python

edades = [18, 22, 21, 19, 24, 23, 20, 22, 21, 25]

edades
````


## Ejercicio 1

Calcular:

- promedio de edades
- edad máxima
- edad mínima


````python
def calcular_promedio(lista):
    return sum(lista) / len(lista)

promedio_edades = calcular_promedio(edades)
print(f"El promedio de edades es: {promedio_edades}")

def calcular_maxima(lista):
    return max(lista)

maxima_edad = calcular_maxima(edades)
print(f"La máxima edad es: {maxima_edad}")

def calcular_minima(lista):
    return min(lista)

minima_edad = calcular_minima(edades)
print(f"La mínima edad es: {minima_edad}")
````


## Ejercicio 2

Contar cuántos estudiantes tienen edad mayor a **21**


````python
def calcular_estudiante_sobre_umbral_edad(lista, umbral):
    estudiantes_sobre_umbral = [edad for edad in lista if edad > umbral]
    return estudiantes_sobre_umbral

umbral_edad = 21
estudiantes_sobre_umbral = calcular_estudiante_sobre_umbral_edad(edades, umbral_edad)
print(f"Estudiantes con edad mayor a {umbral_edad}: {estudiantes_sobre_umbral}")
````


## Ejercicio 3

Crear una función que clasifique edades:

- edad < 21 → "joven"
- edad >= 21 → "adulto"


````python
def clasificador_de_edades(lista):
    clasificacion = []
    for edad in lista:
        if edad < 20:
            clasificacion.append("Joven")
        elif 20 <= edad < 35:
            clasificacion.append("Adulto joven")
        else:
            clasificacion.append("Adulto")
    return clasificacion

clasificacion_edades = clasificador_de_edades(edades)
[print(f"Edad: {edad}, Clasificación: {clasificacion}") for edad, clasificacion in zip(edades, clasificacion_edades)];
````

# 5. Mini desafío

````python

notas = [45, 52, 61, 39, 58, 47, 63, 50]

notas
````


Resolver:

1. Calcular promedio
2. Contar aprobados (> 4.0)
3. Contar reprobados
4. Calcular porcentaje de aprobación


````python
def calcular_promedio(lista):
    return sum(lista) / len(lista)

def contar_aprobados(lista, umbral):
    aprobados = [nota for nota in lista if nota >= umbral]
    return len(aprobados)

def contar_reprobados(lista, umbral):
    reprobados = [nota for nota in lista if nota < umbral]
    return len(reprobados)

def calcular_porcentaje_aprobados(lista, umbral):
    total = len(lista)
    aprobados = contar_aprobados(lista, umbral)
    porcentaje = (aprobados / total) * 100
    return porcentaje

umbral_aprobacion = 55

promedio_notas = calcular_promedio(notas)
aprobados = contar_aprobados(notas, umbral_aprobacion)
reprobados = contar_reprobados(notas, umbral_aprobacion)
porcentaje_aprobados = calcular_porcentaje_aprobados(notas, umbral_aprobacion)

print(f"Promedio de notas: {promedio_notas}")
print(f"Número de aprobados: {aprobados}")
print(f"Número de reprobados: {reprobados}")
print(f"Porcentaje de aprobados: {porcentaje_aprobados:.2f}%")
````

# 6. Repaso de Pandas

Generación de un Dataframe

Guardar datos en distintos formato (csv, parquet, duckdb)

````python
import pandas as pd
import os
import duckdb
#Apaga todos los warnings
import warnings
warnings.filterwarnings("ignore")

data_folder = "data"
if not os.path.exists(data_folder):
    os.makedirs(data_folder)

duckdb_path = os.path.join(data_folder, "estudiantes.duckdb")

data = {
    'Nombre': ['Ana', 'Luis', 'María', 'Carlos', 'Sofía'],
    'Edad': [22, 19, 24, 21, 20],
    'Nota': [85, 78, 92, 88, 90]
}
print(data)

df = pd.DataFrame(data)
display(df)

# Guardar el DataFrame en diferentes formatos

df.to_csv(os.path.join(data_folder, "estudiantes.csv"), index=False)
df.to_parquet(os.path.join(data_folder, "estudiantes.parquet"), index=False)

with duckdb.connect(duckdb_path) as conn:
    df.to_sql("estudiantes", conn, if_exists="replace", index=False)
````

- Cargar datos en diferentes formatos
- Aplicar operaciones sobre los datos
- Filtrar datos según condiciones
- Clasificar datos en categorías

````python
import pandas as pd
import numpy as np

np.random.seed(42)

nombres = [
    "Ana", "Luis", "María", "Carlos", "Sofía", "Pedro", "Valentina", "Diego",
    "Camila", "Mateo", "Fernanda", "Sebastián", "Isabella", "Andrés", "Daniela",
    "Felipe", "Catalina", "Rodrigo", "Javiera", "Nicolás", "Antonia", "Tomás",
    "Constanza", "Ignacio", "Gabriela", "Martín", "Renata", "Benjamín", "Florencia",
    "Agustín"
]

apellidos = [
    "González", "Rodríguez", "Martínez", "López", "García", "Pérez", "Sánchez",
    "Ramírez", "Torres", "Flores", "Rivera", "Morales", "Jiménez", "Vargas", "Castillo",
    "Herrera", "Medina", "Reyes", "Ortiz", "Rojas"
]

carreras = ["Ingeniería Civil", "Ingeniería Informática", "Administración", "Psicología", "Derecho"]

n = 500

df_estudiantes = pd.DataFrame({
    "Nombre": [
        f"{np.random.choice(nombres)} {np.random.choice(apellidos)}"
        for _ in range(n)
    ],
    "Edad": np.random.randint(18, 30, size=n),
    "Nota": np.round(np.random.uniform(30, 100, size=n), 1),
    "Carrera": np.random.choice(carreras, size=n),
    "Semestre": np.random.randint(1, 11, size=n),
    "Creditos_aprobados": np.random.randint(0, 200, size=n),
})

df_estudiantes.head(10)
````

## Exploración inicial del DataFrame

````python
print("--- Dimensiones ---")
print(df_estudiantes.shape)
print(f"Filas: {df_estudiantes.shape[0]}, Columnas: {df_estudiantes.shape[1]}")

print("\n--- Tipos de datos ---")
print(df_estudiantes.dtypes)

n_registros = 3
print(f"\n--- Primeros {n_registros} registros ---")
display(df_estudiantes.head(n_registros))

print(f"\n--- Últimos {n_registros} registros ---")
display(df_estudiantes.tail(n_registros))

top_n_estudiantes = 5
print(f'\n--- Top {top_n_estudiantes} estudiantes con mejores notas ---')
display(df_estudiantes.sort_values(by='Nota', ascending=False).head(top_n_estudiantes))

print(f'\n--- Top {top_n_estudiantes} estudiantes con peores notas ---')
display(df_estudiantes.sort_values(by='Nota', ascending=True).head(top_n_estudiantes))
````

## Estadísticas descriptivas

````python
# describe() entrega un resumen estadístico de las columnas numéricas
df_estudiantes.describe()
````

## Filtrado de datos

````python
# Filtrar estudiantes aprobados (Nota >= 55)
nota_filtrado = 55

condicion = df_estudiantes['Nota'] >= nota_filtrado
aprobados = df_estudiantes[condicion]
print(f"Aprobados: {len(aprobados)} de {len(df_estudiantes)} estudiantes")
display(aprobados.head())

# Filtrar con múltiples condiciones: aprobados de Ingeniería Informática
# Primero generar el array de condiciones con los booleanos

carrera_filtrado = "Ingeniería Informática"

condiciones = (df_estudiantes["Nota"] >= nota_filtrado) & (df_estudiantes["Carrera"] == carrera_filtrado)
aprobados_info = df_estudiantes[condiciones]

print(f"\nAprobados en {carrera_filtrado}: {len(aprobados_info)}")
display(aprobados_info.head())
````

## Columnas calculadas

````python
# Crear columna de estado académico

# # Forma con funcion lambda: 
# df_estudiantes["Estado"] = df_estudiantes["Nota"].apply(
#     lambda nota: "Aprobado" if nota >= 55 else "Reprobado"
# )

# Forma con función tradicional loc:
display(df_estudiantes.head(3))

df_estudiantes["Estado"] = ''
condicion_aprobacion = df_estudiantes["Nota"] >= 55

df_estudiantes.loc[condicion_aprobacion, "Estado"] = "Aprobado"
df_estudiantes.loc[~condicion_aprobacion, "Estado"] = "Reprobado"
display(df_estudiantes.head(3))

# Crear columna de clasificación por nota
def clasificar_nota(nota):
    if nota >= 90:
        return "Distinción"
    elif nota >= 70:
        return "Bueno"
    elif nota >= 55:
        return "Suficiente"
    else:
        return "Reprobado"

df_estudiantes["Clasificacion"] = df_estudiantes["Nota"].apply(clasificar_nota)

df_estudiantes.head(10)
````

## Agrupación con groupby

````python
# Promedio de nota y edad por carrera
resumen_carrera = df_estudiantes.groupby(["Carrera"]).agg(
    Promedio_Nota=("Nota", "mean"),
    Edad_Promedio=("Edad", "mean"),
    Total_Estudiantes=("Nombre", "count"),
    Creditos_Promedio=("Creditos_aprobados", "mean"),
).round(2)

resumen_carrera
````

````python
# Conteo de estudiantes por Estado y Carrera (tabla cruzada)
pd.crosstab(df_estudiantes["Carrera"], df_estudiantes["Estado"])
````

## Ordenamiento y ranking

````python
# Top 10 estudiantes con mejor nota
top10 = df_estudiantes.sort_values("Nota", ascending=False).head(10)
display(top10[["Nombre", "Carrera", "Semestre", "Nota", "Clasificacion"]])

# Ranking dentro de cada carrera
df_estudiantes["Ranking_Carrera"] = (
    df_estudiantes.groupby("Carrera")["Nota"]
    .rank(ascending=False, method="min")
    .astype(int)
)

# Ver los mejores de cada carrera
mejores_por_carrera = (
    df_estudiantes[df_estudiantes["Ranking_Carrera"] == 1]
    [["Nombre", "Carrera", "Nota"]]
    .sort_values("Carrera")
)
print("\nMejor estudiante por carrera:")
display(mejores_por_carrera)
````

## Visualización con matplotlib

````python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 3, figsize=(16, 4))

# 1. Histograma de notas
axes[0][0].hist(df_estudiantes["Nota"], bins=15, color="steelblue", edgecolor="white")
axes[0][0].axvline(55, color="red", linestyle="--", label="Umbral aprobación")
axes[0][0].set_title("Distribución de Notas")
axes[0][0].set_xlabel("Nota")
axes[0][0].set_ylabel("Frecuencia")
axes[0][0].legend()

# 2. Promedio de nota por carrera (barras horizontales)
prom_carrera = df_estudiantes.groupby("Carrera")["Nota"].mean().sort_values()
axes[1][0].barh(prom_carrera.index, prom_carrera.values, color="coral", edgecolor="white")
axes[1][0].axvline(55, color="red", linestyle="--", alpha=0.7)
axes[1][0].set_title("Promedio de Nota por Carrera")
axes[1][0].set_xlabel("Nota Promedio")

# 3. Distribución de clasificaciones (pastel)
conteo_clasificacion = df_estudiantes["Clasificacion"].value_counts()
colors = ["#2ecc71", "#3498db", "#f39c12", "#e74c3c"]
axes[1][1].pie(
    conteo_clasificacion.values,
    labels=conteo_clasificacion.index,
    autopct="%1.1f%%",
    colors=colors[:len(conteo_clasificacion)],
    startangle=90,
)
axes[1][1].set_title("Clasificación Académica")

plt.tight_layout()
plt.show()
````
