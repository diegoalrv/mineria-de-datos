## 1. Cargar dataset

```python
import pandas as pd

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
df.head()
```

## 2. Exploración inicial

```python
print('Shape:', df.shape)
print('\nColumnas:', df.columns.tolist())
df.info()
```

## 3. Estadísticas básicas

```python
df.describe()
```

## 4. Valores faltantes

```python
print('Valores faltantes por columna:')
print(df.isnull().sum())
```

## 5. Ejercicios

### Ejercicio 1 — Edad promedio

```python
print('Edad promedio:', df['Age'].mean())
```

### Ejercicio 2 — Cantidad de hombres y mujeres

```python
print(df['Sex'].value_counts())
```

### Ejercicio 3 — Tasa de supervivencia

```python
tasa = df['Survived'].mean()
print(f'Tasa de supervivencia: {tasa:.2%}')
```

### Ejercicio 4 — Supervivencia por sexo

```python
supervivencia_sexo = df.groupby('Sex')['Survived'].mean()
print('Tasa de supervivencia por sexo:')
print(supervivencia_sexo)
```
